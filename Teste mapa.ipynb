```python
import pandas as pd
import folium
from folium import plugins
```

```python
df = pd.read_csv('arquivos/Coordenadas geogr√°ficas dos CEPs.csv')
df.head()
```

```python
mapa = folium.Map(location=[-3.77, -38.53],titles='Teste de mapa',zoom_start=12)

pontos = df.values.tolist()
for cep,lat,long in pontos:
    folium.Marker([lat,long],popup=cep).add_to(mapa)

mapa
```

```python
mapa = folium.Map(location=[-3.77, -38.53],tiles='OpenStreetMap',zoom_start=12)

pontos = df.values.tolist()
pontos2 = [[b,c] for [a,b,c] in pontos]
mapa.add_child(plugins.HeatMap(pontos2))

mapa
```
